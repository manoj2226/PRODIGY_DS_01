```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data to test your setup
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female'],
    'Age': [25, 30, 22, 35, 28, 24]
}

df = pd.DataFrame(data)

# Show the table
print(df)

# Plot a bar chart of gender
sns.countplot(data=df, x='Gender', palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df = pd.read_csv("Task1_dataset.csv.csv", skiprows=4)
df.head()
```

```python
# Filter data for India and Population
india_pop = df[(df['Country Name'] == 'India') & (df['Indicator Name'] == 'Population, total')]

# Transpose year columns into rows
india_pop = india_pop.loc[:, '1960':'2020'].T  # Select year columns only
india_pop = india_pop.reset_index()
india_pop.columns = ['Year', 'Population']

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=india_pop, x='Year', y='Population', marker='o', color='green')
plt.title("India's Total Population Over Time")
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
selected = df[(df['Indicator Name'] == 'Population, total') & 
              (df['Country Name'].isin(['India', 'China', 'United States']))]

pop_2020 = selected[['Country Name', '2020']]

sns.barplot(data=pop_2020, x='Country Name', y='2020', palette='Set2')
plt.title('Population in 2020')
plt.ylabel('Population')
plt.xlabel('Country')
plt.show()
```

```python
top5_2020 = df[(df['Indicator Name'] == 'Population, total')][['Country Name', '2020']]
top5_2020 = top5_2020.sort_values(by='2020', ascending=False).head(5)

sns.barplot(data=top5_2020, x='Country Name', y='2020', palette='viridis')
plt.title('Top 5 Most Populous Countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
countries = ['India', 'China', 'United States']
pop_data = df[(df['Indicator Name'] == 'Population, total') & df['Country Name'].isin(countries)]

# Transpose years and plot each country
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = pop_data[pop_data['Country Name'] == country].loc[:, '1960':'2020'].T
    country_data.columns = ['Population']
    country_data = country_data.reset_index().rename(columns={'index': 'Year'})
    sns.lineplot(data=country_data, x='Year', y='Population', label=country)

plt.title('Population Growth Comparison (1960â€“2020)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
